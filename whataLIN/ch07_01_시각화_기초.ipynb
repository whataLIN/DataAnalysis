### 여러 개의 선 그어보기

```python
import numpy as np

t=np.arange(0.,5.,0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에서 여러 개의 선')
plt.plot(
    t, t, 'r--',
    t, t ** 2, 'bs:',
    t, t ** 3, 'g^-',
    t, t ** 4, 'y*-.',
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐 표현')
plt.plot((np.arange(4)+1)**2)
plt.plot((np.arange(4)-1)**3)
plt.show()
```

```python
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐 표현')
plt.plot((np.arange(4)+1)**2)
plt.show()      #따로 그리고 싶으면 show 여러번 선언
plt.plot((np.arange(4)-1)**3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)

# 코사인, 사인 곡선
C,S = np.cos(X), np.sin(X)

plt.title("삼각함수 plot")
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sin')
plt.legend(loc=2)  #이름표가 붙어 있다는 뜻임. loc로 위치 설정가능
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치

plt.subplot(2,1,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")

plt.subplot(2,1,2)
plt.plot(np.arange(10), np.arange(10)**3, 'r--')
plt.title("차트2")

# 겹치는 문제 발생
plt.tight_layout()  #안겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치

plt.subplot(1,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")

plt.subplot(1,2,2)    #행과 열을 변경하면 옆으로 그려짐
plt.plot(np.arange(10), np.arange(10)**3, 'r--')
plt.title("차트2")

plt.tight_layout()
plt.show()
```

```python
# 2x2 플롯 배치

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")

plt.xlabel('x축')
plt.ylabel('y축')

plt.subplot(2,2,2)
plt.plot(np.arange(10), np.arange(10)**3, 'r--')
plt.title("차트2")

plt.subplot(2,2,3)
plt.plot(np.arange(10), np.arange(10)**3, 'b*-')
plt.title("차트3")

plt.subplot(2,2,4)
plt.plot(np.arange(10), np.arange(10)**3, 'y^-')
plt.title("차트4")

plt.tight_layout()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y=[2,3,1]
x=np.arange(3)

plt.title("바 차트")

xlabel=['가','나','다'] #len(x)와 길이가 같아야 함
plt.xticks(x,xlabel)  #라벨과 x가 붙음 - x축의 개별 값에 이름 붙이기.

ylabel=[sorted(y)]  #y축에 기준이 되는 이름을 붙여주기.
plt.yticks(sorted(y))

plt.bar(x,y)
plt.show()

#바 차트
```

```python
y=[2,3,1]
x=np.arange(3)

plt.title("가로 차트")

xlabel=['가','나','다'] #len(x)와 길이가 같아야 함
plt.xticks(x,xlabel)  #라벨과 x가 붙음 - x축의 개별 값에 이름 붙이기.

plt.yticks(x,xlabel)

plt.barh(x,y)
plt.show()    #이거다시해
```

```python
people=['james','kate','katii','titi']

y_pos=np.arange(4)
performance=3+10*np.random.rand(4)

error = np.random.rand(4)

plt.title("ber-h chart")

plt.grid(True)
plt.barh(y_pos, performance, xerr = error, alpha=0.4)

plt.yticks(y_pos, people)
plt.xlabel('x축')

plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ["호박파이", "사과파이", "정어리 파이", "호두파이", "엄마손파이"]
sizes=[15,40,45,10,30]
colors=['yellowgreen','gold','lightskyblue','lightcoral','violet']
explode =[0, 0.1, 0, 0, 0.3]
plt.title('파이 차트')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis=('equal')
plt.show()
```
