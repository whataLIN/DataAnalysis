````python
import random

def average_python(n):

  '''
  randomí•œ 0-1ì˜ ì‹¤ìˆ˜ë¥¼ í‰ê· í•´ì¤Œ
  '''

  s=0
  for i in range(n):
    s+=random.random() #0,1ì‚¬ì´ì˜ randomí•œ ìˆ˜
  return s/n
````

````python
%time average_python(100_000)     #%time : ë§¤ì§ë©”ì†Œë“œ. ì‹¤í–‰ì— ê±¸ë¦¬ëŠ” ì‹œê°„ì„ ì¬ ì¤Œ
````

````python
%timeit average_python(100_000)
````
