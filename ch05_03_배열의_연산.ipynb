## ì •ë ¬
* `sort` : ë°°ì—´ ì•ˆì˜ ì›ì†Œë¥¼ í¬ê¸°ì— ë”°ë¼ ì •ë ¬í•˜ì—¬ ìƒˆë¡œìš´ ë°°ì—´ ìƒì„±

````python
np.sort(a)
#axis=-1, axis=1. ì—´ë“¤ ê°„ ì •ë ¬
````
